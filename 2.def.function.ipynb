```python
# recursive function
# it means that a function calls itself
# this has the benefitt of meaning that you can loop through data to reach a result
# factorial number
def fact(n):
  if n==0:
    return 1
  else:
    return n*fact(n-1)
fact(5)
```

```python
# fibonacci number using recursion
def fibonacci(n):
  #base cases
  if n==0:
    return 0
  elif n==1:
    return 1
  # recursive case
  else:
    return fibonacci(n-1) + fibonacci(n-2)
# example usage
print(fibonacci(10))
```
