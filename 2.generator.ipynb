```python
def square(n):
  for i in range(n):
    return i**2
```

```python
# a generator function in python is defined like a normal function, but whenever it needs to generate a value
# it does so with the yield keyword rather than return. If the body of a def contains yield,
# the function automatically becomes a python genertor function.
```

```python
# this case failed
square(3)
```

```python
# generator
def square(n):
  for i in range(n):
    yield i**2
```

```python
square(3)
```

```python
for i in square(3):
  print(i)
```
